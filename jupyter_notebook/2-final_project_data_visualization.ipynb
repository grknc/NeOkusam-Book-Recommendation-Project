## 1. Sorting the Books

Öncelikle veri setindeki kitapları, belli başlı kırılımlara göre sıralama yapalım.

### 1.1.Top Rated Counts(Popular) Books

Öncelikle kitaplara verilen oylara göre sıralama yapalım. Bakalım sonuçlar nasıl olacak?

```python
vote_columns = ['title','book_authors','book_rating_count','book_review_count','book_rating']
```

```python
top_book_rating_count = df1.sort_values(by='book_rating_count', ascending=False).head(15)
top_book_rating_count[vote_columns]
```

#### Visualization

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10, 10))

color = sns.color_palette("Set2")
ax = sns.barplot(x="book_rating_count", y="title", data=top_book_rating_count, palette=color)

for i in ax.patches:
    ax.text(i.get_width() + .05, i.get_y() + 0.5, str(i.get_width()), fontsize = 10, color = 'k')
    
plt.title('Best 15 Books by Book Rating Counts', weight='bold')
plt.show()
```

Toplamda en fazla oy alan ilk 3 kitap
* Harry Potter and Sorcerer's Stone 
* The Hunger Games
* Twilight

## 2. Ranking the Book Authors

Veri seti içerisindeki kitapları sıraladıktan sonra şimdi sırada yazarlarımız var. Kitap yazarlarını belli değişkenlere göre sıralayalım :)

### 2.1.Top 15 Authors with Most Books

İlk olarak veri seti içerisinde en fazla kitaba sahip olan yazarları sıralayalım.

```python
most_books_author = df1.groupby('book_authors')['title'].count().reset_index().sort_values('title', ascending=False).head(15).set_index('book_authors')
plt.figure(figsize=(15,10))

ax = sns.barplot(most_books_author['title'], most_books_author.index, palette='inferno')
ax.set_title("Top 15 authors with most books")
ax.set_xlabel("Total number of books")
totals = []
for i in ax.patches:
    totals.append(i.get_width())
total = sum(totals)
for i in ax.patches:
    ax.text(i.get_width()+.2, i.get_y()+.2,str(round(i.get_width())), fontsize=15,color='black')
plt.show()
```

```python
df1.groupby('book_authors')['title'].count().max()
```

En fazla kitaba sahip olan yazarlarımız 
* Nora Roberts(59)
* Stephen King(59)
* Dean Koontz(47)

### 2.2.Top 15 Authors with Rating Counts

Toplam oy sayılarına göre yazarları sıraladığımızda en iyi 15 yazarımızı şu şekilde gösterebiliriz:

```python
rating_count_authors=df1.groupby('book_authors')['book_rating_count'].sum().sort_values(ascending=False).head(15)
rating_count_authors = rating_count_authors.reset_index()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10, 10))

color = sns.color_palette("magma")
ax = sns.barplot(x="book_rating_count", y="book_authors", data=rating_count_authors, palette=color)

for i in ax.patches:
    ax.text(i.get_width() + .05, i.get_y() + 0.5, str(i.get_width()), fontsize = 10, color = 'k')
    
plt.title('Top 15 Authors by Total Ratings Count', weight='bold')
plt.show()
```

Toplam oy sayısı baz alındığında en yüksek sayı:

* J.K.Rowling , Mary GrandPre
* Suzanne Collins
* Stephenie Meyer

### 2.3.Top 15 Authors with Text Review Count


```python
text_reviews_authors=df1.groupby('book_authors')['book_review_count'].mean().sort_values(ascending=False).head(15)
text_reviews_authors = text_reviews_authors.reset_index()
```

### Top 5 Authors - Average Rating Levels by Book

Veri setinde yüksek puanlı ve iyi skorlara sahip 5 yazarı belirliyoruz. Sonrasında bu yazarlara ilişkin yeni bir veri seti tanımlıyoruz.

```python
authors = ["Stephen King","J.K. Rowling","Suzanne Collins","J.R.R. Tolkien","Stephenie Meyer"]
```

```python
data1 = df1[df1['book_authors'].isin(authors)]
```

```python
data1.head()
```

Kitaplara verilen ortalama değerlendirme puanlarını "low,medium ve high" olarak 3 gruba ayırıyoruz.

```python
bin_edges = [2.5, 3.5, 4.0, 4.5]
bin_names = ['low', 'medium', 'high']
data1['rating_levels'] = pd.cut(data1['book_rating'], bin_edges, labels=bin_names)
data1.head()
```

```python
pd.DataFrame(data1.groupby('book_authors')['rating_levels'].value_counts())
```

```python
data2 = pd.DataFrame(data1.groupby('book_authors')['rating_levels'].value_counts().unstack())
data2
```

Veri setinde Stephen King romanlarına dair ciddi bir sayı mevcut. Bu sonuca göre yazarın çıkardığı kitapların genel bir çoğunluğunun ortalaması 3.5 üzerinde olduğunu varsayabiliriz.

Suzanne Collins yazarının kitapların ortalama derecelendirmesi 4-4.5 aralığında olduğunu varsayabiliriz. Bu durum veri setindeki kitap okuyan kişilerin daha genç olduğunu gösterebilir mi?

```python
labels = 'low', 'medium', 'high'
explode = (0.1, 0, 0)
fig1, [ax1, ax2, ax3,ax4,ax5] = plt.subplots(1, 5, figsize = (18,5))


ax1.pie(data2.iloc[0], explode=explode, labels=labels, autopct='%1.0f%%', shadow=True, startangle=90)
ax1.set_title('J.K. Rowling')
ax1.axis('equal')


ax2.pie(data2.iloc[1], explode=explode, labels=labels, autopct='%1.0f%%', shadow=True, startangle=90)
ax2.set_title('J.R.R. Tolkien')
ax2.axis('equal')

# George R.R. Martin
ax3.pie(data2.iloc[2], explode=explode, labels=labels, autopct='%1.0f%%', shadow=True, startangle=90)
ax3.set_title('Stephen King')
ax3.axis('equal')

# J.K. Rowling
ax4.pie(data2.iloc[3], explode=explode, labels=labels, autopct='%1.0f%%', shadow=True, startangle=90)
ax4.set_title('Stephenie Meyer')
ax4.axis('equal')

# J.R.R. Tolkien
ax5.pie(data2.iloc[4], explode=explode, labels=labels, autopct='%1.0f%%', shadow=True, startangle=90)
ax5.set_title('Suzanne Collins')
ax5.axis('equal')


plt.show()
```

### The Distribution of Book Ratings 

Veri setindeki kitapların derecelendirmelerine ilişkin durumu gözlemlemek istiyoruz.

Kitaplara verilen puanların yoğunluğunu incelediğimizde ağırlıklı olarak (3,8 - 4.3) değerinde olduğunu görmekteyiz.

### Ratings vs Review Count

```python
plt.figure(figsize=(15,10))
sns.set_context('paper')
ax =sns.jointplot(x="book_rating",y='book_review_count', kind='scatter',  data= df1[['book_review_count', 'book_rating']])
ax.set_axis_labels("Average Rating", "Text Review Count")
plt.show()
```

(3.5-4.5) arasında ortalama derecelendirmeye göre kitapların yorum sayılarının yoğunlukla 8000 civarında olduğunu söyleyebiliriz. O zaman bu değer altındaki yorum sayılarına göre yoğunluğu inceleyelim.

```python
trial = df[~(df['book_review_count']>8000)]
```

```python
plt.figure(figsize=(15,10))
sns.set_context('paper')
ax =sns.jointplot(x="book_rating",y='book_review_count', kind='kde',  data= trial,color="green")
ax.set_axis_labels("Average Rating", "Text Review Count")
plt.show()
```

Bu grafiğe göre 4.5 üzerinde değerlendirme alan kitapların yorum sayılarının her zaman yüksek olacağını söyleyemeyiz.

```python
most_books_author = df.groupby('book_format')['title'].count().reset_index().sort_values('title', ascending=False).head().set_index('book_format')
plt.figure(figsize=(15,8))

ax = sns.barplot(most_books_author['title'], most_books_author.index, palette='Set1')
ax.set_title("Top 5 Book Formats Count",size=14)
ax.set_xlabel("Total number of books",size=10)
ax.set_ylabel("Book Formats",size=10)
totals = []
for i in ax.patches:
    totals.append(i.get_width())
total = sum(totals)
for i in ax.patches:
    ax.text(i.get_width()+.2, i.get_y()+.2,str(round(i.get_width())), fontsize=15,color='black')
plt.show()
```

```python
data3 = df[df['book_format'].isin(special_format)]
```

```python
plt.figure(figsize=(25,10))
sns.lmplot(data = data3, x = "book_rating_count", y = "book_review_count", hue = "book_format",legend=False,palette="Set1",
           height=10, aspect=1.2)
plt.legend(loc='lower right')
plt.title('Rating Count vs Review Count per book pages level', weight='bold',size="14")
plt.show()
```
